```python
import pandas as pd

from tqdm import tqdm
```

```python
data_21 = pd.read_csv('data/atp_matches_2021.csv')
data_22 = pd.read_csv('data/atp_matches_2022.csv')
data_23 = pd.read_csv('data/atp_matches_2023.csv')
data_24 = pd.read_csv('data/atp_matches_2024.csv')

data = pd.concat([data_21, data_22, data_23, data_24], axis=0, ignore_index=True)
```

```python
tourney_level_of_interest = ['A', 'G', 'M', 'F']
data = data[data['tourney_level'].isin(tourney_level_of_interest)].reset_index(drop=True)
data
```

```python
total_accurate_predictions = 0
for _, row in tqdm(data.iterrows()):
    if row.winner_rank_points > row.loser_rank_points:
        total_accurate_predictions += 1
```

```python
"ATP accuracy:", total_accurate_predictions / len(data)
```
